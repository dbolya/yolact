```python
import os, glob, matplotlib.pyplot as plt, moviepy.editor as mp
from tqdm import tqdm
```

```python
files = glob.glob('../../../../datasets/slip-fall/mladen-holdout/05-FPS/1x/*.mp4')
pbar = tqdm(files)
# Write resized videos
for file in pbar:
    pbar.set_description('Working on {}'.format(os.path.basename(file)))
    clip = mp.VideoFileClip(file)
    if clip.rotation in (90, 270):
        clip = clip.resize(clip.size[::-1])
    clip.rotation = 0
    clip = clip.resize(0.25)
    path_out = '../../../../datasets/slip-fall/mladen-holdout/05-FPS/0.25x/'
    clip.write_videofile(os.path.join(path_out,os.path.basename(file)), logger = None)
```

```python
# Sanity check results
import VideoReaders
frame_provider = VideoReaders.VideoReader('../../../../datasets/slip-fall/mladen-holdout/05-FPS/0.25x/0018.mp4')
f,ax = plt.subplots(1,figsize = (10,10))
shoot = []
for frame in frame_provider:
    shoot.append(frame)
ax.imshow(shoot[14][:,:,::-1])
ax.set(title = 'Shape of {}'.format(shoot[14].shape))
```
